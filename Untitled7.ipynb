```python
df = pd.read_csv('iris.data', header=None)
nombres_col=['longitud_sepalo','ancho_sepalo','longitud_petalo','ancho_petalo','Etiqueta']
df.columns=nombres_col
X_cols=['longitud_sepalo','ancho_sepalo','longitud_petalo','ancho_petalo']
df.head()
```

Estandandarizar y nommalizar los datos

```python
# Change 'descomposition' to 'decomposition'
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
df[X_cols] = ss.fit_transform(df[X_cols])

# Assuming 'X' should be 'df[X_cols]' to display the scaled data
df[X_cols].head()
```

Empezamos a programar para PCA

```python
pca2=PCA(n_components=2,random_state=42)
pca_2= pca2.fit_transform(df[X_cols])
pca_2[:6]
```

```python
df_2=pd.DataFrame({'PCA1':pca_2[:,0],'PCA2':pca_2[:,1],'Etiqueta':df['Etiqueta']})
df_2.head()
```

```python
pca2.explained_variance_ratio_
```

```python
pca2.explained_variance_ratio_.sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x=['PCA1','PCA2'], y=pca2.explained_variance_ratio_)
plt.show()

```

```python
sns.scatterplot(data=df_2,x='PCA1',y='PCA2',hue='Etiqueta')
plt.show()
```

```python
pca3=PCA(n_components=3,random_state=42)
pca_3= pca3.fit_transform(df[X_cols])
pca_3[:6]
```

```python
df_3=pd.DataFrame({'PCA1':pca_3[:,0],'PCA2':pca_3[:,1],'PCA3':pca_3[:,2],'Etiqueta':df['Etiqueta']})
df_3.head()
```

```python
pca3.explained_variance_ratio_
```

```python
pca3.explained_variance_ratio_.sum()
```

```python
sns.barplot(x=['PCA1','PCA2','PCA3'], y=pca3.explained_variance_ratio_)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Define a dictionary mapping labels to colors
color_mapping = {
    'Iris-setosa': 'red',
    'Iris-versicolor': 'green',
    'Iris-virginica': 'blue'
}

# Map the 'Etiqueta' column to colors using the dictionary
colors = df_3['Etiqueta'].map(color_mapping)

# Create the 3D scatter plot with the mapped colors
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
x = df_3['PCA1']
y = df_3['PCA2']
z = df_3['PCA3']

# Pass the mapped colors to the 'c' argument
ax.scatter(x, y, z, c=colors)
ax.set_xlabel('PCA1')
ax.set_ylabel('PCA2')
ax.set_zlabel('PCA3')
plt.title('PCA Visualization')

plt.show()
```
